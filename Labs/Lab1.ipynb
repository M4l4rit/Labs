### Задание №2
Написать генератор случайных матриц(многомерных), который принимает
опциональные параметры <b>m</b>, <b>n</b>, <b>min_limit</b>, <b>max_limit</b>, где <b>m</b> и <b>n</b> указывают размер
матрицы, а <b>min_lim</b> и <b>max_lim</b> - минимальное и максимальное значение для
генерируемого числа.

```python
import random
import time

user_m = int(input())
user_n = int(input())
user_min_limit = int(input())
user_max_limit = int(input())
user_m, user_n, user_min_limit, user_max_limit


matrix = []
for i in range(user_n):
    matrix.append([0]*user_m)


for i in range(user_n):
    for j in range(user_m):
        matrix[i][j] = round(random.uniform(user_min_limit, user_max_limit))


for i in range(len(matrix)):         
    for j in range(len(matrix[i])):  
        print(matrix[i][j], end = ' ')
    print() 
```

### Задание №3
Реализовать методы сортировки строк числовой матрицы в соответствии с
заданием. Оценить время работы каждого алгоритма сортировки и сравнить его со
временем стандартной функции сортировки. Испытания проводить на сгенерированных
матрицах.

```python
import copy

# Сортировка выбором.
# Берётся срез массива, в котором минимальный элемент переносят в самый левый угол,
# после чего срез уменьшается и цикл повторяется.

_matrix = copy.deepcopy(matrix)

start_time = time.time()


def sort_selection(matrix):
    for i in range(len(matrix)):
        index = i

        for j in range(i + 1, len(matrix)):
            if matrix[index] > matrix[j]:
                index = j
                
        temp = matrix[index]
        matrix[index] = matrix[i]
        matrix[i] = temp
            

for i in range(len(_matrix)):
    sort_selection(_matrix[i])

sort_selection_time = round((time.time() - start_time)*1000)
print("{0}ms".format(sort_selection_time))

for i in range(len(matrix)):         
    for j in range(len(matrix[i])):  
        print(_matrix[i][j], end = ' ')
    print() 



```

```python
# Сортировка вставкой.

_matrix = copy.deepcopy(matrix)

start_time = time.time()


def sort_insert(matrix):
    for i in range(1, len(matrix)):
        index = matrix[i]
        j = i - 1
        
        while j >= 0 and matrix[j] > index:
            matrix[j + 1] = matrix[j]
            j -= 1

        matrix[j + 1] = index
        
for i in range(len(_matrix)):
    sort_insert(_matrix[i])

sort_insert_time = round((time.time() - start_time)*1000)
print("{0}ms".format(sort_insert_time))
for i in range(len(matrix)):         
    for j in range(len(matrix[i])):  
        print(_matrix[i][j], end = ' ')
    print() 
```

```python
# Сортировка обменом. || Сортировка пузырьком.

_matrix = copy.deepcopy(matrix)

start_time = time.time()


def sort_exchange(matrix):
    for i in range(len(matrix)):
        index = 0
        for j in range(1, len(matrix) - i):
            if matrix[j] < matrix[j - 1]:
                temp = matrix[j]
                matrix[j] = matrix[j - 1]
                matrix[j - 1] = temp
                index += 1
                
        if index == 0:
            break
            
        
for i in range(len(_matrix)):
    sort_exchange(_matrix[i])

sort_exchange_time = round((time.time() - start_time)*1000)
print("{0}ms".format(sort_exchange_time))
for i in range(len(matrix)):         
    for j in range(len(matrix[i])):  
        print(_matrix[i][j], end = ' ')
    print() 
```

```python
# Сортировка Шелла.

_matrix = copy.deepcopy(matrix)

start_time = time.time()


def sort_shell(matrix):
    index = int(len(matrix) / 2)
    while index > 0:
        for i in range(index, len(matrix), index):
            key = matrix[i]
            j = i - index
            
            while j >= 0 and key < matrix[j]:
                matrix[j + index] = matrix[j]
                j -= index
                
            matrix[j + index] = key
        index = int(index / 2)
        
        
for i in range(len(_matrix)):
    sort_shell(_matrix[i])

sort_shell_time = round((time.time() - start_time)*1000)
print("{0}ms".format(sort_shell_time))
for i in range(len(matrix)):         
    for j in range(len(matrix[i])):  
        print(_matrix[i][j], end = ' ')
    print() 
```

```python
# Быстрая сортировка.

_matrix = copy.deepcopy(matrix)

start_time = time.time()


def partition(matrix, first, last):
    _matrix = matrix[first]
    left = first + 1
    right = last

    while True:
        while matrix[right] > _matrix and right > first:
            right -= 1

        while matrix[left] <= _matrix and left < last:
            left += 1

        if left >= right:
            break

        temp = matrix[left]
        matrix[left] = matrix[right]
        matrix[right] = temp

    matrix[first] = matrix[right]
    matrix[right] = _matrix

    return right


def _quick_sort(matrix, first, last):
    if last > first:
        pivot_index = partition(matrix, first, last)
        _quick_sort(matrix, first, pivot_index - 1)
        _quick_sort(matrix, pivot_index + 1, last)
   

def quick_sort(matrix):
    _quick_sort(matrix, 0, len(matrix) - 1)
        
        
for i in range(len(_matrix)):
    quick_sort(_matrix[i])
        
quick_sort_time = round((time.time() - start_time)*1000)
print("{0}ms".format(quick_sort_time))
for i in range(len(matrix)):         
    for j in range(len(matrix[i])):  
        print(_matrix[i][j], end = ' ')
    print() 
```

```python
# Турнирная сортировка.

import heapq

_matrix = copy.deepcopy(matrix)

start_time = time.time()

def tournament_sort(matrix, g):
    _g = []
    if len(matrix) >= g:
        _g = matrix[:g]
    else:
        _g = matrix[:]
    heapq.heapify(_g)
    
    winners = [float('-inf')]
    losers = []
    
    i = g
    while i < len(matrix) or len(_g) > 0:
        top = heapq.heappop(_g)
        if top >= winners[-1]:
            winners.append(top)
        else:
            losers.append(top)
        
        if i < len(matrix):
            heapq.heappush(_g, matrix[i])
        i += 1
    

    del winners[0]
    matrix.clear()
    matrix.extend(losers)
    matrix.extend(winners)
    

    if len(losers) > 0:
        tournament_sort(matrix, g)
    

for i in range(len(_matrix)):
    tournament_sort(_matrix[i], 10)


tournament_sort_time = round((time.time() - start_time)*1000)
print("{0}ms".format(tournament_sort_time))
for i in range(len(matrix)):         
    for j in range(len(matrix[i])):  
        print(_matrix[i][j], end = ' ')
    print() 
```
